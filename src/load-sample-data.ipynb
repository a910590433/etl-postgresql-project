##### **Load sample data**

```python
import requests
import urllib3
from io import StringIO
import pandas as pd
import os

url_110 = "https://opdadm.moi.gov.tw/api/v1/no-auth/resource/api/dataset/8736F690-4B8E-43FE-8CFE-1335109BB3D4/resource/91F53D99-29E3-490B-A3D1-1ABA8CCA3F12/download"

response = requests.get(url_110, verify=False)
df_sample = pd.read_csv(StringIO(response.text))

# Since SSL certificates for some government data portals may be missing or invalid, suppress the InsecureRequestWarning for this personal project.
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

folder_path = r"C:\Users\user\Downloads\etl-first-marriage\data"

# Full file path
file_path = os.path.join(folder_path, "sample_data.csv")

# Save a random 5-row sample
df_sample.sample(5).to_csv(file_path, index=False)

print("Sample data has been saved")
```
